# Pandas Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Windows 10 PRO\Downloads\Ice Cream Ratings.csv")
df = df.set_index('Date')
```

```python
df
```

```python
print(plt.style.available)
plt.style.use('fivethirtyeight')
```

```python
df.plot(kind = 'line', subplots = True)
```

```python
df.plot(kind = 'line', title = 'Icecream ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
df['Flavor Rating'].plot(kind = 'bar', stacked = True)
```

```python
df.plot.barh(stacked = True)
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 200, c = 'Orange')
```

```python
df.plot.hist(bins = 10)
```

```python
df.boxplot()
```

```python
df.plot.area(figsize = (10,5))
```

```python
df.plot.pie(y = 'Flavor Rating', figsize = (10,6))
```
